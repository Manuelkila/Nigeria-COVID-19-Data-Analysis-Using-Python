
## Data Cleaning

```python
#A function that converts all non int data to int. 
def int_converter(x):
    while type(x) != int:
        return int(x.replace(',', ''))
        
        
```

```python
#applying the int_converter function on the Nigeria's Covid-19 data
nigCo['No. of Cases (Lab Confirmed)']= nigCo['No. of Cases (Lab Confirmed)'].apply(int_converter)
nigCo['No. of Cases (on admission)'] = nigCo['No. of Cases (on admission)'].apply(int_converter)
nigCo['No. Discharged'] = nigCo['No. Discharged'].apply(int_converter)
#nigCo['No. of Deaths'] = nigCo['No. of Deaths'].apply(int_converter)
```

```python
#Rename columns
nigCo.columns = ['states', 'cases_lab_confirmed', 'cases_on_admission', 'discharged', 'deaths']
```

```python
#A function that extracts Nigeria's data from world's data
def get_nig_data(df):
    df = df.drop(['Province/State', 'Lat', 'Long'], axis=1)
    nig = df.groupby('Country/Region').sum().loc['Nigeria'].to_frame().reset_index()
    nig.columns =  ['Date', 'Cases']
    nig['Date'] = pd.to_datetime(nig['Date'])
    return nig
    
```

```python
#applying get_nig_data to extract Nigeria's deaths, confirmed and recovered data
nig_deaths = get_nig_data(df_deaths)
nig_recovered = get_nig_data(df_recovered)
nig_confirmed = get_nig_data(df_confirmed)
```

## Analysis and Visualization 

```python
#A function for plotting top ten states for each columns
def largest(column_name, title, ax):
    data = nigCo.sort_values(column_name, ascending = False).head(10)
    plt.figure(figsize=(8,6))
    ax = sns.barplot(x=column_name, y='states', data=data[['states', column_name]], ax=ax, palette = 'rainbow')
    ax.set_title(title, fontsize =20)
    #plt.tight_layout()
    return ax
```

```python
nigCo['deaths_per_100']= round((nigCo['deaths']/nigCo['cases_lab_confirmed'])*100)
```

```python
nigCo[['states','deaths_per_100']].sort_values('deaths_per_100', ascending = False)
```

```python
fig, ax = plt.subplots(2, 3, figsize=(25, 13))
largest('deaths','States with Highest Death Cases', ax=ax[0][2])
largest('cases_on_admission', "Top Active Cases", ax=ax[0][1])
largest('discharged', 'Discharged Cases', ax=ax[1][0])
largest('cases_lab_confirmed', 'Confirmed Cases by Laboratory', ax=ax[0][0])
#largest('deaths_per_100', 'Number of Deaths per 100 Cases', ax = ax[1][1])



pll =nigCo[nigCo['states']!='Kogi'].nlargest(10,'deaths_per_100')
plt.figure(figsize=(8,6))
sns.barplot(x='deaths_per_100', y='states', data=pll[['states', 'deaths_per_100']], palette = 'rainbow', ax = ax[1][1])
ax[1][1].set_title('Number of Deaths Per 100 Cases', fontsize =16)


ax[1][2].axis('off')
plt.tight_layout()
fig.savefig('seven.png', dpi=400, bbox_inches ='tight')
```

```python
#Line plot function
def trends(data, title, ax, color):
    ax= sns.lineplot(x="Date", y="Cases", data=data, ax=ax, color=color)
    #xticks(rotation=35)
    ax.set_title(title, fontsize =20,fontweight="bold" )
    return ax
```

```python
fig, axarr = plt.subplots(2, 2, figsize=(20, 15))
trends(nig_recovered,'Total Daily Recovered cases in Nigeria',ax=axarr[0][1], color='green')
trends(nig_confirmed,'Total Daily Confirmed Cases in Nigeria',ax=axarr[0][0], color = 'blue')
trends(nig_deaths,'Total Daily Death cases in Nigeria',ax=axarr[1][0], color='red')
y = nig_confirmed['Cases'].diff()
sns.lineplot(x="Date", y=y, data=nig_confirmed, ax=axarr[1][1], color ='blue')
axarr[1][1].set_title('Daily Infection rate in Nigeria', fontsize =20,fontweight="bold")
plt.tight_layout()
plt.savefig('two.png', dpi=500, bbox_inches ='tight')
```

- From the boxplot, we can see that most states have less than 1000 confirmed cases. if we consider the correlation plot above, we'll see that there is no strong correlation. It is possible that the low number of confirmed cases in most states is too small to determine any clear relationship.  

```python
df_join_10 = df_join.sort_values('cases_lab_confirmed', ascending = False).head(12)
```

```python
(df_join_10['cases_lab_confirmed'].sum()/df_join['cases_lab_confirmed'].sum())*100
```

- The top 12 states by number of confirmed cases account for 80.5 percent of the total confrimed cases.  Let us determine if there are corelations between the top 12 states and CCVI indexes. 

```python
def reg_plot(title,x, ax):
    data = df_join_10
    ax= sns.regplot(x=x, y='cases_lab_confirmed', data=data, ax=ax)
    #xticks(rotation=35)
    ax.set_title(title, fontsize =22, fontweight="bold")
    return ax
```

```python
fig, axarr = plt.subplots(3, 3, figsize=(25, 13))
reg_plot('Confirmed Cases VS Population density', 'Population Density', ax =axarr[0][0])
reg_plot('Age VS Population density', 'Age', ax =axarr[0][1])
reg_plot('Confirmed Cases vs Epidemiological', 'Epidemiological',ax=axarr[0][2])
reg_plot('Confirmed Cases vs Fragility', 'Fragility',ax=axarr[1][0])
reg_plot('Confirmed Cases vs Health System', 'Health System',ax=axarr[1][1])
reg_plot('Confirmed Cases vs Socio-Economic', 'Socio-Economic',ax=axarr[1][2])
reg_plot('Confirmed Cases vs  Transport Availability', ' Transport Availability',ax=axarr[2][0])
reg_plot('Confirmed Cases vs  Acute IHR', 'Acute IHR',ax=axarr[2][1])
reg_plot('Confirmed Cases vs  Overall CCVI Index', 'Overall CCVI Index',ax=axarr[2][2])

plt.tight_layout()
plt.savefig('one.png', dpi=500, bbox_inches ='tight')
```

- The chart above shows strong correlation between number of confirmed cases and population density, epidemiological index and transport availability. Epidemilogical index is highly related to population density, so the similar trend in the two measures is expected. 

```python
df_join_10[['cases_lab_confirmed', 'Population Density', 'Epidemiological',' Transport Availability']].corr('pearson')
```

### Impact on Nigeria's Economy

```python
df_join2 = pd.merge(df_join, df_budget, how='outer', on=['states'])
```

```python
df_join2['changes']= df_join2['Initial_budget (Bn)']- df_join2['Revised_budget (Bn)']
df_join2['% decrease']= ((df_join2['Initial_budget (Bn)']- df_join2['Revised_budget (Bn)'])/df_join2['Initial_budget (Bn)'])*100
```

```python
df_join2[['states','changes','% decrease']].nlargest(10,'changes')
```

```python
#total reduction in budget
(df_join2['changes'].sum()/df_join2['Initial_budget (Bn)'].sum())*100
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 8))
#unique = df_join2['states'].unique()
#palette = dict(zip(unique, sns.color_palette(n_colors=len(unique))))
sns.barplot(x='% decrease', y='states',data=df_join2.sort_values('% decrease',ascending = False).head(10), 
            ax =ax[2], palette='rainbow')

sns.barplot(x='changes', y='states',data=
            df_join2.sort_values('changes',ascending = False).head(10), ax=ax[1], palette='rainbow')


sns.barplot(x='Initial_budget (Bn)',y='states',data = df_join2.nlargest(10,'Initial_budget (Bn)'),
            palette='magma', ax=ax[0])
ax[1].set_title('Decrease in Budget by States', fontsize =16)


ax[2].set_title('Percentage Decrease in Budget by States', fontsize=16)
ax[0].set_title('Top Ten States by Budget', fontsize=16)
plt.tight_layout()
plt.savefig('five.png', dpi=500, bbox_inches ='tight')
```

```python
dfgroup = df_join2.groupby('region')['Initial_budget (Bn)', 'Revised_budget (Bn)', 'cases_lab_confirmed'].sum().reset_index()

dfgroup['changes'] = abs(dfgroup['Revised_budget (Bn)'] - dfgroup['Initial_budget (Bn)'])

dfgroup = dfgroup.sort_values('changes', ascending = False)
dfgroup['percent_decrease'] = (dfgroup['changes']/dfgroup['Initial_budget (Bn)'])*100
```

```python
dfgroup
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8))
unique = dfgroup['region'].unique()
palette = dict(zip(unique, sns.color_palette(n_colors=len(unique))))
sns.barplot(x='percent_decrease', y='region',data=dfgroup.sort_values('percent_decrease',ascending = False), 
            ax =ax[1], palette=palette)
sns.barplot(x='changes', y='region',data=dfgroup, ax=ax[0], palette=palette)
ax[0].set_title('Decrease in Budget by Region', fontsize =20)
ax[1].set_title('Percentage Decrease in Budget by Region', fontsize=16)
plt.tight_layout()
plt.savefig('four.png', dpi=500, bbox_inches ='tight')
```

- Although, the north central region recorded the highest decrease in budget, the south west region actually had the highest percentage decrease in budject, which further butress the impact of the lockdown on the region.
